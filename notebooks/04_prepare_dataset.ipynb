```python
import shutil
import os

def copy_images(df, split_name):

    copied = 0
    missing = []

    for _, row in df.iterrows():

        src = f"../data/processed/{row['platform']}/{row['file_name']}"
        dst = f"../data/final/{split_name}/{row['platform']}/{row['file_name']}"

        try:
            shutil.copy(src, dst)
            copied += 1

        except Exception as e:
            missing.append((src, str(e)))

    print("Copied:", copied)
    print("Errors:", len(missing))

    return missing
```

```python
errors = copy_images(train_df, "train")
errors[:5]
```

```python
import os

valid_rows = []

for _, row in clean_df.iterrows():

    path = f"../data/processed/{row['platform']}/{row['file_name']}"

    if os.path.exists(path):
        valid_rows.append(row)

valid_df = pd.DataFrame(valid_rows)

print("Original:", len(clean_df))
print("Valid files:", len(valid_df))
print("Removed:", len(clean_df) - len(valid_df))
```

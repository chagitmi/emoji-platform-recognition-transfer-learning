```python
import os
import base64

selected_platforms = [
    "Apple",
    "Google",
    "Facebook",
    "Windows",
    "Twitter",
    "JoyPixels",
    "Samsung"
]

# נבדוק רק 3 שורות ראשונות
sample_df = df.head(3)

for index, row in sample_df.iterrows():

    emoji_name = row["name"].replace(" ", "_").replace("/", "_")

    for platform in selected_platforms:

        # אם אין תמונה - לדלג
        if pd.isna(row[platform]):
            continue

        try:
            # לקחת string
            image_data = row[platform]

            # להפריד header
            _, encoded = image_data.split(",", 1)

            # decode
            image_bytes = base64.b64decode(encoded)

            # שם קובץ
            filename = f"{emoji_name}_{index}.png"

            # נתיב
            save_path = f"../data/processed/{platform}/{filename}"

            # לשמור
            with open(save_path, "wb") as f:
                f.write(image_bytes)

        except Exception as e:
            print(f"Error at row {index}, {platform}: {e}")

print("Done")
```
